# IceCube Event Viewer — example notebook

This notebook shows the three main ways to use the viewer:

1. **Load from SQLite** — the most common workflow
2. **Load from DataFrames** — when you already have the pulses in memory
3. **Save to file** — export as HTML or image

```python
# If the package is not yet installed, install it from the repo root:
# !pip install -e ..

from icecube_event_viewer import show_sql_event, show_event, query_sql
```

## 1 · Load from SQLite

```python
DB_PATH = "/path/to/your/database.db"  # <- change this
EVENT_NO = 12345                        # <- change this

fig, truth, event = show_sql_event(
    db_path=DB_PATH,
    event_no=EVENT_NO,
    truth_cols=["energy", "zenith", "azimuth", "classification"],
    mode="show",
    show_direction_arrow=True,
    show_first_hit=True,
)

print(truth)
```

## 4 · Custom camera angle

```python
import numpy as np

def camera_from_angle(deg: float, z: float = 1.0, distance: float = 2.0) -> dict:
    rad = np.radians(deg)
    return dict(x=distance * np.cos(rad), y=distance * np.sin(rad), z=z)

# fig.update_layout(scene_camera=dict(eye=camera_from_angle(45)))
# fig.show()
```
